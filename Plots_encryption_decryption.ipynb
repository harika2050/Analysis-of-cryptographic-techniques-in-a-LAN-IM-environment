```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 5
aes= (0.0842, 0.13, 0.2344, 0.3206, 0.3933)
des = (17.36, 150.81, 310.18, 571.17, 1758.13)
rsa= (1.54, 15.94, 36.99, 80.97, 325.84)
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, aes, bar_width,
alpha=opacity,
color='b',
label='AES')

rects2 = plt.bar(index + bar_width, des, bar_width,
alpha=opacity,
color='g',
label='DES')
rects2 = plt.bar(index + bar_width+bar_width, rsa, bar_width,
alpha=opacity,
color='r',
label='RSA')

plt.xlabel('File Size')
plt.ylabel('Encryption time in seconds')
plt.title('Encryption Times')
plt.xticks(index + bar_width, ('50KB', '500KB', '1MB', '2MB', '5MB'))
plt.legend()
ax.set_ylim([0,40])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 5
aes= (41, 43, 44.2, 47.9, 58.1)
des = (41.4, 42.4, 44.6, 48.2, 52.4)
rsa= (45.8, 76.3, 110.2, 150.4, 351.9)
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
y_pos1= (0, 4, 8, 12, 16)
y_pos2=(1, 5, 9, 13, 17)
y_pos3= (2, 6, 10, 14, 18)
y_pos= (1,5,9,13,17)
rects1 = plt.bar(y_pos1, aes, bar_width,
alpha=opacity,
color='b',
label='AES')

rects2 = plt.bar(y_pos2, des, bar_width,
alpha=opacity,
color='g',
label='DES')
rects2 = plt.bar(y_pos3, rsa, bar_width,
alpha=opacity,
color='r',
label='RSA')

plt.xlabel('File Size')
plt.ylabel('Memory Utilisation in MiB')
plt.title('Comparison of memory utilisation during encryption')
plt.xticks(y_pos, ('50KB', '500KB', '1MB', '2MB', '5MB'))
plt.legend()
ax.set_ylim([0,200])
plt.tight_layout()
plt.show()
```
